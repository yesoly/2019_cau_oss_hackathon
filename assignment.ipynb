```
# 사용할 라이브러리 import
import torch
import torch.nn as nn # 선형 회귀 모델 함수
import torch.nn.functional as func

torch.manual_seed(1)

# y = 3x+2 의 임의의 일차함수를 잡고, x, y 데이터 학습 데이터 만들기
x_train = torch.FloatTensor([[1],[2],[3]])
y_train = torch.FloatTensor([[5],[8],[11]])

# Linear(1,1) : 하나의 입력에 하나의 출력
model = nn.Linear(1,1)

# SGD : 경사하강법, lr : learning rate, model.parameters()를 사용하여 W와 b를 전달
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# 경사 하강법 2000번 반복
nb_epochs = 2000
for epoch in range(nb_epochs+1):

    # y값 예측
    prediction = model(x_train)

    # cost 계산, mse_loss : 파이토치의 평균오차제곱 함수
    cost = func.mse_loss(prediction, y_train)

    # gradient를 0으로 초기화
    optimizer.zero_grad()

    # cost를 미분해서 0에 가까워질 때 오차가 최소가 된다.
    cost.backward()

    # W와 b를 업데이트
    optimizer.step()

    if epoch % 100 == 0:
    # 100번마다 로그 출력
      print('Epoch {:4d}/{} Cost: {:.6f}'.format(epoch, nb_epochs, cost.item()))

# 학습을 통해 구한 W와 b의 값
print(list(model.parameters()))

# 임의의 숫자를 넣었을 때 예측 값
test_x =  torch.FloatTensor([[8.0]]) 
pred_y = model(test_x)

print("입력이", test_x ,"일때의 예측값 :", pred_y) 
```
